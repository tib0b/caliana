```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import caliana
print('caliana', caliana.__version__)
```

```python
# load qt, which will be used to display the interface
%gui qt
```

```python
# create a session object from a .tif file, with no sub-sampling.
# here we load from the synthetic .tif file used for testing purpose. Select your own file for the actual analysis.
# Note that s is the "session object", which contains all the variables and functions used during analysis.
s = caliana.Session.from_file('../tests/data/synthetic_calcium_imaging.tif', temporal_step=1)
print('stack shape (T, Y, X):', s.data.shape)
```

```python
# Optional: set frame interval, here for example one frame every 10 seconds.
s.set_frame_interval(10)
```

```python
# Show a scrollable preview of the dataset, and a heatmap of the max normalized intensity.
s.preview()
```

```python
# Optional: Draw boxes on the leaves to serve as reference for leaf movement tracking.
# The boxes should only contain leaf tissue (no background).
# You can place as many leaf boxes as you want, but registration time will increase with the number and size of the boxes.
s.select_leaves()
if s.leaf_regions:
    s.register(caliana.RegistrationMode.PER_LEAF, reference='previous')
```

```python
# Select the ROIs for the analysis. You can also use the free-hand roi tool to select a whole leaf, for example.
s.select_rois()
```

```python
# Optional: preview the ROI traces and crop them to a time interval.
# Drag the shaded window (or type start/end frames), then "Crop to window".
# The interval is stored on the session so analyze() sees the same window.
traces = s.crop_traces()
```

```python
# Open the analysis screen, analysis is split into two tabs :
# - "Traces analysis" is where you can analyze the traces extracted from the ROIs
# - "Heatmaps" lets you view the entire dataset as a heatmap
s.analyze()
```
